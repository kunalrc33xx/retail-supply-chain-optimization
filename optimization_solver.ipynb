```python
# ==========================================
# PART 1: DATA LOADING & SETUP
# ==========================================

# 1. Load Raw Data
# Warehouse Data: [ID, Rent, Capacity]
wh_raw = [
    [0, 6200, 1600], [1, 8000, 1600], [2, 7000, 1200], [3, 5800, 1200],
    [4, 5000, 1400], [5, 6800, 1000], [6, 5200, 1400], [7, 5600, 1400],
    [8, 6200, 1200], [9, 5800, 1400]
]

# Store Data: [ID, Intercept, Slope, Rent]
store_raw = [
    [0, 272.51, -4.13], [1, 326.45, -5.48], [2, 273.92, -3.5], [3, 297.73, -5.63],
    [4, 277.85, -5.36], [5, 302.4, -4.3], [6, 308.34, -4.22], [7, 324.67, -4.24],
    [8, 375.37, -3.95], [9, 376.94, -4.79], [10, 325.68, -4.56], [11, 296.16, -5.26],
    [12, 379.2, -4.28], [13, 333.74, -4.77], [14, 336.75, -4.4], [15, 334.91, -4.77],
    [16, 328.58, -5.11], [17, 281.17, -4.6], [18, 315.08, -5.02], [19, 260.93, -4.11],
    [20, 305.88, -4.17], [21, 257.52, -4.73], [22, 293.08, -5.0], [23, 307.04, -4.52],
    [24, 266.32, -5.1], [25, 272.42, -4.3], [26, 310.7, -5.52], [27, 239.21, -4.89],
    [28, 272.01, -4.48], [29, 263.19, -4.7], [30, 282.83, -4.53], [31, 326.94, -4.65],
    [32, 263.53, -4.87], [33, 306.69, -5.71], [34, 327.17, -4.57], [35, 321.78, -5.14],
    [36, 295.19, -4.85], [37, 283.88, -6.23], [38, 362.6, -5.04], [39, 310.39, -3.91]
]

# Load Travel Time CSVs (Required for Phase 3)
try:
    whtt_df = pd.read_csv('whtt.csv')
    storett_df = pd.read_csv('storett.csv')
    print("Travel time files loaded successfully.")
except FileNotFoundError:
    print("Error: 'whtt.csv' or 'storett.csv' not found. Routing calculation will fail.")

# 2. Define Abstract Sets and Parameters
I = [f"Store{row[0]}" for row in store_raw]  # Set of Stores
J = [f"WH{row[0]}" for row in wh_raw]        # Set of Warehouses

# Calculate Demand (Phase 1 Logic: Price $37.20)
PRICE = 37.20
D = {f"Store{row[0]}": round(row[1] + row[2] * PRICE, 2) for row in store_raw}
K = {f"WH{row[0]}": row[2] for row in wh_raw} # Capacity
F = {f"WH{row[0]}": row[1] for row in wh_raw} # Rent

# ==========================================
# PART 2: OPTIMIZATION MODEL (GUROBI)
# ==========================================

print("\n--- Solving Phase 2: Warehouse Location ---")
try:
    m = gp.Model("Warehouse_Optimization")
    m.setParam('OutputFlag', 0) # Silence Gurobi output

    # Decision Variables
    y = m.addVars(J, vtype=GRB.BINARY, name="Open")
    x = m.addVars(I, J, vtype=GRB.CONTINUOUS, name="Ship")

    # Objective: Minimize Fixed Rent
    m.setObjective(gp.quicksum(F[j] * y[j] for j in J), GRB.MINIMIZE)

    # Constraint 1: Demand Satisfaction
    m.addConstrs((gp.quicksum(x[i, j] for j in J) == D[i] for i in I), name="Demand")

    # Constraint 2: Capacity
    m.addConstrs((gp.quicksum(x[i, j] for i in I) <= K[j] * y[j] for j in J), name="Capacity")

    m.optimize()

    if m.status == GRB.OPTIMAL:
        print(f"Optimal Solution Found. Total Rent: ${m.objVal:,.2f}")
    else:
        print("Optimization failed.")
        exit()

except gp.GurobiError as e:
    print(f"Gurobi Error: {e}")
    exit()

# ==========================================
# PART 3: ROUTING LOGIC (HEURISTIC)
# ==========================================

# Routing Constants
VEHICLE_CAPACITY = 500
MAX_ROUTE_TIME = 480
UNLOAD_TIME = 30
VEHICLE_FIXED_COST = 200
VEHICLE_VAR_COST = 0.1

def get_travel_time(from_loc, to_loc, is_from_wh=False):
    """Fetch travel time from loaded DataFrames."""
    to_id = int(to_loc.replace('Store', ''))
    if is_from_wh:
        wh_row_id = from_loc.lower()
        col_name = f"store{to_id}"
        row = whtt_df[whtt_df['wh'] == wh_row_id]
        if not row.empty: return row.iloc[0][col_name]
    else:
        from_id = int(from_loc.replace('Store', ''))
        row = storett_df[storett_df['store'] == f"store{from_id}"]
        if not row.empty: return row.iloc[0][f"store{to_id}"]
    return float('inf')

def generate_routes(wh_id, assigned_stores):
    """Nearest Neighbor Heuristic."""
    routes = []
    unvisited = set(assigned_stores)

    while unvisited:
        # Start new route
        start_node = min(unvisited, key=lambda s: get_travel_time(wh_id, s, is_from_wh=True))
        current_route = [start_node]
        current_load = D[start_node]

        # Initial Time: Travel from WH + Unload
        travel_to = get_travel_time(wh_id, start_node, is_from_wh=True)
        current_time = travel_to + UNLOAD_TIME

        curr_node = start_node
        unvisited.remove(start_node)

        # Add neighbors
        while unvisited:
            candidates = sorted(list(unvisited), key=lambda s: get_travel_time(curr_node, s))
            added = False
            for next_node in candidates:
                dist = get_travel_time(curr_node, next_node)
                return_dist = get_travel_time(wh_id, next_node, is_from_wh=True)

                new_load = current_load + D[next_node]
                new_time = current_time + dist + UNLOAD_TIME + return_dist

                if new_load <= VEHICLE_CAPACITY and new_time <= MAX_ROUTE_TIME:
                    current_route.append(next_node)
                    current_load = new_load
                    current_time += dist + UNLOAD_TIME
                    curr_node = next_node
                    unvisited.remove(next_node)
                    added = True
                    break
            if not added: break

        # Finalize Route
        return_time = get_travel_time(wh_id, curr_node, is_from_wh=True)
        total_time = current_time + return_time
        cost = VEHICLE_FIXED_COST + (total_time * VEHICLE_VAR_COST)

        routes.append({'path': current_route, 'cost': cost, 'time': total_time, 'load': current_load})

    return routes

# ==========================================
# PART 4: EXECUTION & RESULTS
# ==========================================

print("\n--- Phase 3: Generating Routes ---")
total_logistics_cost = 0
final_routes_output = []

for j in J:
    if y[j].X > 0.5:
        # Identify stores assigned to this warehouse by Gurobi
        assigned = [i for i in I if x[i, j].X > 0.01]

        # Generate Routes
        wh_routes = generate_routes(j, assigned)

        print(f"\nWarehouse {j} (Assigned: {len(assigned)} stores):")
        for r in wh_routes:
            path_str = " -> ".join(r['path'])
            print(f"  Route: {path_str}")
            print(f"    Load: {r['load']:.2f} | Time: {r['time']:.1f} min | Cost: ${r['cost']:.2f}")
            total_logistics_cost += r['cost']

            # Format for routes.txt
            wh_clean = j.replace("WH", "")
            segments = [f"{s.replace('Store', '')}:{D[s]:.2f}" for s in r['path']]
            final_routes_output.append(f"{wh_clean}," + ",".join(segments))

print("-" * 40)
print(f"Total Logistics Cost: ${total_logistics_cost:,.2f}")
print(f"Total Fixed Rent:     ${m.objVal:,.2f}")

# Final Profit Calc
GROSS_PROFIT = 82219.22
final_profit = GROSS_PROFIT - m.objVal - total_logistics_cost
print(f"Final Estimated Profit: ${final_profit:,.2f}")
```
