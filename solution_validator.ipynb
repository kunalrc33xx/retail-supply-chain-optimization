```python
#@title Define other input data
vehcap = 500
vehfixedcost = 200 # Fixed cost of a vehicle
vehvarcost = 0.1 # Variable cost of driving a vehicle; $/minute
maxtime = 480 # Maximum length permitted for route
unitcost = 10 # Unit cost at warehouse
maxpricechange = 0.05 # maximum price difference per minute travel time
stoptime = 30 # Time it requires to unload product at a store, in minutes
```

```python
#@title Functions that parse solution files.
class Route(typing.NamedTuple):
    wh: int
    stops: tuple[tuple[int,float],...]

class Solution(typing.NamedTuple):
    openwhs: set[int]
    openstores: set[int]
    whstock: dict[int, float]
    rtes: list[Route]
    prices: dict[int,float]

def read_wh_file(whfilename):
    with open(whfilename, 'r') as f:
      openwhs = set()
      whstock = {}
      for line in f:
        splitline = line.split(',')
        wh = int(splitline[0])
        stock = float(splitline[1])
        openwhs.add(wh)
        whstock[wh] = stock
    return openwhs, whstock

def read_store_file(storefilename):
    with open(storefilename, 'r') as f:
      openstores = set()
      storestock = {}
      prices = {}
      for line in f:
        splitline = line.split(',')
        store = int(splitline[0])
        price = float(splitline[1])
        openstores.add(store)
        prices[store] = price
    return openstores, prices

def read_rte_file(rtefilename):
   with open(rtefilename, 'r') as f:
      rtes = []
      for line in f:
         splitline = line.split(',')
         wh = int(splitline[0])
         stops = []
         for stop in splitline[1:]:
            splitstop = stop.split(':')
            store = int(splitstop[0])
            q = float(splitstop[1])
            stops.append((store, q))
         rtes.append(Route(wh, stops))
   return rtes
```

```python
#@title Functions that calculate objective and check feasibility.
def calc_rte_traveltime(route:Route, whtts, storetts, stoptime):
    """
    This calculates the travel time cost incurred by a route.
    wh: an integer; id of warehouse where route starts and ends
    stores: list of ints, the stores visited on the route
    whtts: matrix; whdists[i, j] gives travel time from warehouse i to store j
    storetts: matrix; storedists[i, j] gives travel time from store i to store j
    """
    stores = [stop[0] for stop in route.stops]
    totaltime = whtts[route.wh, stores[0]] + stoptime * len(route.stops)
    for i in range(1,len(stores)-1):
        totaltime += storetts[stores[i],stores[i+1]]
    totaltime += whtts[route.wh,stores[-1]]
    return totaltime

def calc_obj(sol: Solution, intercepts, slopes,
             whfixed, storefixed, vehfixed,
             whvar, vehvar, whtt, storett, stoptime):
  obj = -sum(whfixed[wh] for wh in sol.openwhs)
  obj -= sum(storefixed[wh] for wh in sol.openstores)
  obj -= sum(whvar * prodq for prodq in sol.whstock.values())
  storestock = {s: 0.0 for s in sol.openstores}
  for rte in sol.rtes:
    obj -= vehfixed
    obj -= calc_rte_traveltime(rte, whtt,storett,stoptime)*vehvar
    for (store, quantity) in rte.stops:
        storestock[store] += quantity
  for s in sol.openstores:
    obj += sol.prices[s]*min(intercepts[s] + slopes[s] * sol.prices[s],
                             storestock[s])

  return obj

def check_transport_constr(sol: Solution, vehcap, vehmaxtime, whtts, storetts, stoptime):
    violated = False
    whship = {w: 0.0 for w in sol.openwhs}
    for i,r in enumerate(sol.rtes):
        total_time = calc_rte_traveltime(r,whtts,storetts, stoptime)
        if total_time > vehmaxtime:
            violated = True
            warnings.warn(f"Violated constraint: {i}th route requires {total_time} minutes to complete, which is more than the maximum permitted time of {vehmaxtime} minutes.")
        total_transported = 0
        if r.wh not in sol.openwhs:
            violated = True
            warnings.warn(f"Violated constraint: {i}th route starts at warehouse {r.wh}, which was not listed as an opened warehouse. Routes should only start at open warehouses.")
        for (store, q) in r.stops:
            if store not in sol.openstores:
                violated = True
                warnings.warn(f"Violated constraint: {i}th route stops at store {store}, which was not listed as an opened store. Routes should only stop at open stores.")
            total_transported += q
            if r.wh in sol.openwhs:
                whship[r.wh] += q

        if total_transported > vehcap:
            violated = True
            warnings.warn(f"Violated constraint: {i}th route transports a total of {total_transported} units; this shouldn't be higher than the vehicle capacity of {vehcap} units.")

    for wh in sol.openwhs:
        if whship[wh] > sol.whstock[wh]:
            violated = True
            warnings.warn(f"Violated constraint: total amount transported from warehouse {wh} is {whship[wh]} while stock at warehouse {wh} is listed as {sol.whstock[wh]}; warehouse stock should be at least amount transported from warehouse.")
    return violated

def check_price_constr(sol: Solution, storetts, maxpricechange):
    for s1 in sol.openstores:
      for s2 in sol.openstores:
        if abs(sol.prices[s1] - sol.prices[s2]) > maxpricechange*storetts[s1,s2]:
          warnings.warn(f"Violated constraint: price change between store {s1} and {s2} exceeds maximum allowable. Note: other pairs of stores may also violate this constraint.")
          return True
    return False
```

```python
# Read in solution files
openwhs, whstock = read_wh_file('warehouse.txt')
openstores, prices = read_store_file('store.txt')
rtes = read_rte_file('route.txt')
whtt = pandas.read_csv('whtt.csv', index_col=["wh"]).to_numpy()
sol = Solution(openwhs, openstores, whstock, rtes, prices)

# Read in data files.
storett = pandas.read_csv('storett.csv', index_col=["store"]).to_numpy()
storedata = pandas.read_csv('storedata.csv', index_col=["store"])
whdata = pandas.read_csv('whdata.csv', index_col=["wh"])

# Check for feasibility
constr1viol = check_transport_constr(sol, vehcap, maxtime, whtt, storett, stoptime)
constr2viol = check_price_constr(sol,storett,maxpricechange)

# If feasible, calculate profit.
if not (constr1viol or constr2viol):
  obj = calc_obj(sol, storedata["inter"].to_numpy(), storedata["slope"].to_numpy(),
         whdata["rent"].to_numpy(), storedata["rent"].to_numpy(), vehfixedcost,
             unitcost, vehvarcost, whtt, storett, stoptime)
  print(f"Solution is feasible. Profit achieved: {obj}")
else:
  print(f"Solution is infeasible.")
```
